```python
import numpy as np
from tensorflow.keras.datasets import cifar10

# 載入CIFAR-10資料集
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
# 輸出資料形狀
print('X_train:', X_train.shape, 'y_train:', y_train.shape)
print('X_test :', X_test.shape, 'y_test :', y_test.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 繪製圖像
num_classes = 10          #種類的數量
pos = 1                  #維持圖像位置的變數
#有幾個類別，就重複幾次

for target_class in range(num_classes):
# 維持分類到各個類別中的圖像索引的清單 
    target_idx = [] 
# 類別i為答對時，則取得正確答案標籤 
    for i in range(len(y_train)): # i列、0行的標籤是否與target_class一致 
        if y_train[i][0] == target_class: # 類別i為答對時，則將標籤追加到targetIdx
            target_idx.append(i)

    np.random.shuffle(target_idx)    # 洗牌類別i的圖像索引
    plt.figure(figsize=(20, 20))      # 將繪圖區域設定為橫向20英吋、直向20英吋

    # 將最初洗牌的10張圖像描繪出來
    for idx in target_idx[:10]:
        plt.subplot(10, 10, pos)        # 指定繪圖區域為10列、10行的第pos項的位置
        plt.imshow(X_train[idx])       # 使用Matplotlib的imshow()進行繪圖
        pos += 1

plt.show()
```
